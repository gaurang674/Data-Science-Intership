```python
import pandas as pd
```

### Read the loan_data_set.csv file and perform the cleaning of it.

```python
loan = pd.read_csv("C:/Users/Gaurang Vaghela/OneDrive/Desktop/Dataset/loan_data_set.csv")
loan
```

```python
loan.isnull().sum()
```

```python
loan.ffill(inplace=True)
loan.bfill(inplace=True)
```

```python
loan = pd.get_dummies(loan, columns=['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status'])
```

```python
loan.dtypes
```

```python
loan.drop('Gender_Female',axis=1,inplace=True)
loan.drop('Married_No',axis=1,inplace=True)
loan.drop('Dependents_3+',axis=1,inplace=True)
loan.drop('Education_Graduate',axis=1,inplace=True)
loan.drop('Self_Employed_No',axis=1,inplace=True)
loan.drop('Property_Area_Rural',axis=1,inplace=True)
loan.drop('Loan_Status_N',axis=1,inplace=True)
```

```python
loan
```

### Identify the type of ML model suitable for the data set. Hint: based on the nature of data set best model will be Logistic Regression.

```python
x=loan.iloc[:,2:15]
y=loan.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=2)
    
ml = LogisticRegression()
ml.fit(x_train,y_train)
y_pred = ml.predict(x_test)

y_pred
```

```python
y_test
```

### After application of model find the accuracy of model on the dataset along with confusion matrix.

```python
from sklearn.metrics import confusion_matrix, accuracy_score
acc = accuracy_score(y_test,y_pred)
conf = confusion_matrix(y_test,y_pred)
print(acc)
print(conf)
```
